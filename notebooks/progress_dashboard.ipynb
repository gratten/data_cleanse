```python
df = filterx.sld(df_og)
```

```python
all_files = df['name_low'].value_counts(ascending=True).rename_axis('unique_values').reset_index(name='count')
```

```python
mult1 = all_files['count'].value_counts().index.to_list()
occur1 = all_files['count'].value_counts().to_list()

for i in range(len(mult1)):
    occur1[i] = occur1[i] * mult1[i]
```

```python
# total prt and asm files
totalFiles = len(df)

# total duplicates
totalDups = len(df) - occur1[0]
```

```python
df_dups_singles = pd.DataFrame({'occurences':occur1}, index=mult1)
# totalFiles = str(len(all_files))

# title = fileName + titlestring + totalFiles
title = f'{fileName}\nTotal files: {totalFiles}\n Total dups: {totalDups}'
p = df_dups_singles.plot.pie(y='occurences', figsize=(10, 10), title=title)
```

```python
dups = df[df.duplicated(['name_low'], keep=False)]
dups = dups.sort_values(['name_low'])
dups = dups.reset_index(drop=True)
```

```python
all_dups = dups['name_low'].value_counts(ascending=True).rename_axis('unique_values').reset_index(name='count')
```

```python
mult_list = all_dups['count'].value_counts().index.to_list()
occur_list = all_dups['count'].value_counts().to_list()

for i in range(len(mult_list)):
    occur_list[i] = occur_list[i] * mult_list[i]
```

```python
df_dups = pd.DataFrame({'occurences':occur_list}, index=mult_list)
totalFiles = str(len(all_dups))

pie.pie(df_dups, fileName, totalFiles, 3)
```
